```python
from tkinter import *
import timeit
import random as rnd

def partition(arr, ini, fin):
    i = ini - 1
    pivote = arr[fin][1]
    
    for j in range(ini,fin):
        if arr[j][1] > pivote:
            i = i + 1
            arr[i],arr[j] = arr[j],arr[i]
    arr[i+1],arr[fin] = arr[fin],arr[i+1]
    return i+1

def quicksort(arr, ini, fin):
    n = len(arr)
    
    if ini < fin:
        pi = partition(arr, ini, fin)
        
        quicksort(arr, ini, pi-1)
        quicksort(arr, pi+1, fin)

###################################################

def nfdhSPP(arrItem, anchBin, altoBin):
    n = len(arrItem)
    quicksort(arrItem, 0, n-1)
    bins = 0
    y = altoBin
    
    
    ########
    arrItem[0] = (arrItem[0][0], arrItem[0][1], 0, y - arrItem[0][1], bins)
    newEst = arrItem[0][3]
    for i in range(1, n):
        varx = arrItem[i-1][2] + arrItem[i-1][0] + arrItem[i][0]
        
        if varx <= anchBin: ## Si el ancho del item es menor o igual al anchoBin, agregar
            arrItem[i] = (arrItem[i][0], arrItem[i][1], arrItem[i-1][2] + arrItem[i-1][0], y - arrItem[i][1], bins)
        else:
            y = newEst ## Actualiza posicion y para un nuevo estante
            vary = y - arrItem[i][1]
            
            if vary >= 0: ## Si el alto del item es mayor o igual a 0, agregar
                newEst = newEst - arrItem[i][1]
                arrItem[i] = (arrItem[i][0], arrItem[i][1], 0, vary, bins)
            else:
                newEst = arrItem[i][3]
                y = altoBin
                bins = bins + 1 ## Agregamos una plancha nueva

                arrItem[i] = (arrItem[i][0], arrItem[i][1], 0, y - arrItem[i][1], bins)
                
    return arrItem, bins


##INI
###############################################################################
bins = 0

anchBin = 150
altoBin = 150
numItem = 10


arrItem = [(0,0,0,0,0)]*numItem

for i in range(numItem):
    anchItem = rnd.randint(30, 80) 
    altItem = rnd.randint(30, 80) 
    arrItem[i] = (anchItem, altItem, 0, 0, 0)

arrItem, bins = nfdhSPP(arrItem, anchBin, altoBin)
print(arrItem)
print("Planchas utilizadas: ", bins)
print("---------------------")
print("Tiempo de ejecucion: ")
%timeit nfdhSPP(arrItem, anchBin, altoBin)

##BIN
###############################################################################
binn = Tk()
binn.title('canvas')

canvas = Canvas(width=400, height=400, bg='white')
canvas.pack(expand=YES, fill=BOTH)


for i in range(bins):
    canvas.create_rectangle((i*anchBin)+0, 0, i*anchBin+anchBin, altoBin, width=4, fill='white')
for b in range(numItem):
    canvas.create_rectangle((anchBin*arrItem[b][4])+arrItem[b][2], arrItem[b][3], (anchBin*arrItem[b][4])+arrItem[b][2]+arrItem[b][0], (arrItem[b][3]+arrItem[b][1]), width=4, fill='green')


binn.mainloop()
 
```
